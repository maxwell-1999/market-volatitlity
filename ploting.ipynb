```python
import csv
import math
def rawSD(close_prices):
    mean = sum(close_prices) / len(close_prices)
    diff = [(close_prices[i] - mean) ** 2 for i in range(len(close_prices))]
    varience = sum(diff) / (len(close_prices))
    sd = math.sqrt(varience)
    return sd
def round_dec(num):
    return round(num*100,4)
```
